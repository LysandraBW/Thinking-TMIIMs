```python
import pandas as pd
```

```python
df_0 = pd.read_excel('0_Papers.xlsx')
print(df_0.shape)
df_0.head()
```

```python
df_1 = pd.read_csv('1_Papers_Cleaned.csv')
print(df_1.shape)
df_1.head()
```

```python
df_2 = pd.read_csv('2_Papers_Screened.csv')
print(df_2.shape)
df_2.head()
```

```python
df_3 = pd.read_csv('3_Papers_Screened.csv')
print(df_3.shape)
df_3.head()
```

```python
import pickle
```

```python
with open('4-ScreenByEntities-0.pickle', 'rb') as file:
    mask = pickle.load(file)
print(f'Keys: {mask.keys()}')
print(f'Mask Length: {len(mask["mask"])}')
```

```python
df_4 = df_3[mask['mask']]
print(df_4.shape)
df_4.head()
```

```python
df_1_dimensions_urls = df_1['Dimensions URL'].tolist()
df_2_dimensions_urls = df_2['Dimensions URL'].tolist()
df_3_dimensions_urls = df_3['Dimensions URL'].tolist()
df_4_dimensions_urls = df_4['Dimensions URL'].tolist()
```

```python
df_0_filtered = df_0[~df_0['Dimensions URL'].isin(df_1_dimensions_urls)]
print(df_0_filtered.shape)
assert df_0_filtered.shape[0] == df_0.shape[0] - df_1.shape[0]

df_0_filtered_sample = df_0_filtered.sample(n=20)
df_0_filtered_sample.to_csv('0_Papers_Cleaned_And_Excluded_By_Topic_Sample.csv', index=False)
```

```python
df_1_filtered = df_1[~df_1['Dimensions URL'].isin(df_2_dimensions_urls)]
print(df_1_filtered.shape)
assert df_1_filtered.shape[0] == df_1.shape[0] - df_2.shape[0]

df_1_filtered_sample = df_1_filtered.sample(n=20)
df_1_filtered_sample.to_csv('1_Papers_Excluded_By_Language_Sample.csv', index=False)
```

```python
df_2_filtered = df_2[~df_2['Dimensions URL'].isin(df_3_dimensions_urls)]
print(df_2_filtered.shape)
assert df_2_filtered.shape[0] == df_2.shape[0] - df_3.shape[0]

df_2_filtered_sample = df_2_filtered.sample(n=20)
df_2_filtered_sample.to_csv('2_Papers_Excluded_By_Keyword_Sample.csv', index=False)
```

```python
df_3_filtered = df_3[~df_3['Dimensions URL'].isin(df_4_dimensions_urls)]
print(df_3_filtered.shape)
assert df_3_filtered.shape[0] == df_3.shape[0] - df_4.shape[0]

df_3_filtered_sample = df_3_filtered.sample(n=20)
df_3_filtered_sample.to_csv('3_Papers_Excluded_By_Entity_Sample.csv', index=False)
```
