```python
import time
import pandas as pd
from langdetect import detect
from IPython.display import clear_output
```

```python
df = pd.read_csv("CleanedPapers.csv")
print(f"Shape: {df.shape}")
```

```python
# # Testing
# df = pd.read_excel("Benchmark-03-09.xlsx")
# print(f"Shape: {df.shape}")
```

```python
mask = []
errors = []
counts = {
    None: 0
}

t0 = time.time()
texts = df.Abstract.tolist()

for i, text in enumerate(texts):
    try:
        lang = detect(text)
        mask.append(lang == "en")
        counts[lang] = counts.get(lang, 0) + 1
    except Exception as e:
        errors.append((i, e))
        mask.append(None)
        counts[None] += 1

    t1 = time.time()
    elapsed = int(t1 - t0)
    m, s = divmod(elapsed, 60)
    h, m = divmod(m, 60)
    
    clear_output(wait=True)
    print(f"{i+1}/{len(texts)} ({h:d}:{m:02d}:{s:02d})")
    print(f"Number Included: {counts['en']}")
    print(f"Number Excluded: {sum([count for lang, count in counts.items() if lang != 'en'])}")
    print(f"Number Errors: {counts[None]}")
```

```python
# Save Screen Data
df[mask].to_csv("ScreenedPapers-2.csv", index=False, encoding="utf-8")
```
