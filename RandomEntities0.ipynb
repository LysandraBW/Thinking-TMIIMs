```python
def is_abbreviated(name: str) -> Union[str, Tuple[str, str]]:
    if not re.search(r"[A-Za-z]\.\s[A-Z-a-z]+", name):
        return False
    return name.split()

def find_scientific_name_query(name: Union[str, Tuple[str, str]], only_species=True):
    where_species = "" if not only_species else 'AND NameUsage."col:rank" = \'species\''
    where = None
    if isinstance(name, str):
        where = f'WHERE NameUsage."col:scientificName" LIKE \'{name}\''
    else:
        where = f'WHERE SUBSTR(NameUsage."col:genericName", 1, 1) LIKE \'{name[0][0]}\' AND NameUsage."col:specificEpithet" LIKE \'{name[1]}\''
    
    return f"""
        SELECT *
        FROM NameUsage
        {where}
        {where_species}
    """

def find_vernacular_name_query(name: str, only_english=False):
    where_english = "" if not only_english else 'AND VernacularName."col:language" = \'eng\''
    
    return f"""
        SELECT *
        FROM VernacularName
        WHERE VernacularName."col:name" LIKE '{name}'
        {where_english}
    """

def find_scientific_name_subs_query(name: Union[str, Tuple[str, str]], only_species=True, only_english=False):
    where = None
    if isinstance(name, str):
        where = f'WHERE NameUsage."col:scientificName" LIKE \'{name}\''
    else:
        where = f'WHERE SUBSTR(NameUsage."col:genericName", 1, 1) LIKE \'{name[0][0]}\' AND NameUsage."col:specificEpithet" LIKE \'{name[1]}\''

    where_species = "" if not only_species else 'AND NameUsage."col:rank" = \'species\''
    where_english = "" if not only_english else 'AND VernacularName."col:language" = \'eng\''
    
    return f"""
        SELECT *
        FROM NameUsage
        JOIN VernacularName 
        ON NameUsage."col:ID" = VernacularName."col:taxonID"
        {where}
        {where_species}
        {where_english}
    """

def find_vernacular_name_subs_query(name: str, only_species=True, only_english=False):
    where_species = "" if not only_species else 'AND NameUsage."col:rank" = \'species\''
    where_english = "" if not only_english else 'AND VernacularName."col:language" = \'eng\''
    
    return f"""
        SELECT NameUsage.*, VernacularName.*
        FROM NameUsage
        JOIN VernacularName 
        ON NameUsage."col:ID" = VernacularName."col:taxonID"
        WHERE VernacularName."col:name" LIKE '{name}'
        {where_species}
        {where_english}
    """

def search_scientific_name(name: str) -> DataFrame:
    binomial = is_abbreviated(name) or name
    query = find_scientific_name_query(binomial)
    print(query)
    return pd.read_sql(query, conn)

def search_vernacular_name(name: str) -> DataFrame:
    query = find_vernacular_name_query(name)
    return pd.read_sql(query, conn)

def find_scientific_name_substitutions(name: str) -> list[str]:
    name = is_abbreviated(name) or name
    query = find_scientific_name_subs_query(name)
    output = pd.read_sql(query, conn)
    
    # Get Substitutions
    output = output["col:name"].tolist()
    output = [text.lower() for text in output]
    output = [*set(output)]
    
    return output
    
def find_vernacular_name_substitutions(name: str):
    query = find_vernacular_name_subs_query(name)
    output = pd.read_sql(query, conn)

    # Get Substitutions
    output = search_name(name)
    output = output["col:scientificName"].tolist()
    output = [text.lower() for text in output]
    output = [*set(output)]
    
    return output
```

```python
import re
from rich.console import Console

console = Console()

def highlight(string, substrings):
    for substring in substrings:
        string = re.sub(f"({re.escape(substring)})", r"[bold black on bright_yellow]\1[/bold black on bright_yellow]", string, flags=re.IGNORECASE)
    console.print(string)

highlight("The quick brown fox jumps over the lazy dog", ["fox", "dog"])
```
